Bethel and Will -- I chose these features based on a pretty loose logic:  We need the basic ID stuff of course, like the first few features, and then the demographic/physical stuff, like age/height/weight/gender.  But it's pretty much a free-for-all after that.  I relied a lot on the criteria they used for that DNN model feature selection in the article I linked to in our DM.  Specifically, the features that leak info about how old the app user is won't help us much.  So e.g. categories like "how old are your children" kind of gives away the user's age.  Or "how long have you lived at current address", or a ton of other info, is just very correlated with user age.  So why does that even matter?  I'm not sure :) , because our ML experts in this project haven't clearly specified their target variables, which is a debate I can't wait to hear, but at any rate, our features should be useful in determining how much someone's BA/RoA/BAA, whatever you call it, is changing, based on their actions.  So I'm kind of avoiding things that reveal the person's actual age ("CA"), other than including their age here so that we can check back on them later in the MIDUS study and see "Oh, look, they've aged 8 years on the calendar, but their survey responses are more like 10 years of BA".  

But even claiming I used any logic here, I was not consistent in some of the columns I included/omitted.  For example, a lot of broad categories have annoyingly many questions to ask the same thing ("Do you feel like you are in control of your life?", "Do you feel like someone else is controlling your life", "If you set your mind to something can you accomplish it", "Has everything already been determined already", etc, etc).  So in some cases, I included all 10 annoying questions, but you'll find that there's usually a sort of summary number that MIDUS used to quantify it all in one number, at the end of the section.  So I included a mix of those types here, and maybe you guys can weigh in on what you think.
